# Не забудьте, что атрибут добавляется при помощи декоратора

```python
class Aircraft:
	def __init__(self, weight):
		self._weight = weight

class UAV:
  def __init__(self):
    self._has_autopilot = True
    self._missions = []

  @property
  def missions(self):
    return self._missions

  def add_mission(self, mission):
    self._missions.append(mission) 
  
  def count_missions(self):
    return len(self._missions)

class MultirotorUAV(Aircraft, UAV):
  def __init__(self, weight, model, brand):
    super().__init__(weight)
    UAV.__init__(self)
    self.__weight = weight
    self.__model = model
    self.__brand = brand
    self.__incidents = []

  def get_info(self):
    return f"Информация о дроне {self.__model}: масса {self.__weight}, производитель {self.__brand}, количество миссий {self.count_missions()}"
  
  def get_model(self):
    return self.__model

  @property
  def incidents(self):
    return self.__incidents;

  def add_incident(self, incident):
    self.incidents.append(incident)
  
  def save_data(self):
    # Название файла
    file_name = self.__class__.__name__ + self.get_model() + ".json"
    # Заполнение блоков файла
    data={}
    data['model'] = self.__model
    data['weight'] = self.__weight
    data['brand'] = self.__brand
    data['missions'] = self.missions
    data['incidents'] = self.__incidents
    # Запись в файл
    with open("/content/drive/MyDrive/Colab Notebooks/" + file_name, 'w') as outfile:
      json.dump(data, outfile, ensure_ascii=True, indent = 2)
```

Создаем экземляры класса

```python
drone_catalog = {
	"DJI Mavic 2 Pro": {"weight":903, "brand":"DJI"},
	"DJI Mavic 2 Zoom": {"weight":900, "brand":"DJI"},
	"DJI Mavic 2 Enterprise Advanced": {"weight":920, "brand":"DJI"},
	"DJI Inspire 2": {"weight":1500, "brand":"DJI"},
	"DJI Mavic 3": {"weight":1000, "brand":"DJI"}
}
#Записываем экземпляры класса в список
Drones = []
for drone in drone_catalog:
  Drone = MultirotorUAV(drone_catalog[drone]['weight'], drone, drone_catalog[drone]['brand'])
  Drones.append(Drone)
#Присваиваем экземплярам класса миссии
pilots = json.load(open ("/content/drive/MyDrive/Colab Notebooks/pilot_path.json"))
for pilot in pilots:
  for mission in pilots[pilot]['missions']:
    model = mission['drone']
    for miss in mission['mission']:
      for drone in Drones:
        if model==drone.get_model():
          drone.add_mission(miss)
```
